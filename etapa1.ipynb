```python
import pandas as pd

# 1. Cargar el dataset original
df = pd.read_csv('datosFormulario.csv')

# 2. Filtrar el dataset para quedarnos SOLO con los estudiantes
# Nota: Ajustá el texto entre comillas si en tu CSV dice exactamente otra cosa (ej: 'Estudiante')
df_estudiantes = df[df['Condicion_estudiante'].str.contains('Estudiante', case=False, na=False)]

# Guardamos la columna de interés ya filtrada
redes_est = df_estudiantes['Horas_redes_dia_promedio']
total_est = len(df_estudiantes)

print(f"Total de estudiantes reales encontrados en la muestra: {total_est}\n")
print("=== RESPUESTAS EXACTAS PARA EL PUNTO E (SOLO ESTUDIANTES) ===")

# 1. Cantidad entre 3 y 8 horas
c1 = redes_est[(redes_est >= 3) & (redes_est <= 8)].count()
print(f"• Cantidad de estudiantes entre 3 y 8 horas diarias: {c1} estudiantes.")

# 2. Porcentaje más de 6 horas
c2 = (redes_est[redes_est > 6].count() / total_est) * 100
print(f"• Porcentaje de estudiantes que usan más de 6 horas: {round(c2, 2)}%")

# 3. Cantidad menos de 2 horas
c3 = redes_est[redes_est < 2].count()
print(f"• Cantidad de estudiantes que usan menos de 2 horas: {c3} estudiantes.")

# 4. Porcentaje entre 5 y 10 horas
c4 = (redes_est[(redes_est >= 5) & (redes_est <= 10)].count() / total_est) * 100
print(f"• Porcentaje de estudiantes entre 5 y 10 horas: {round(c4, 2)}%")

# 5. Cantidad más de 8 horas
c5 = redes_est[redes_est > 8].count()
print(f"• Cantidad de estudiantes que usan más de 8 horas: {c5} estudiantes.")

# 6. Porcentaje menos de 4 horas
c6 = (redes_est[redes_est < 4].count() / total_est) * 100
print(f"• Porcentaje de estudiantes que usan menos de 4 horas: {round(c6, 2)}%")

# 7. Cantidad entre 1 y 6 horas
c7 = redes_est[(redes_est >= 1) & (redes_est <= 6)].count()
print(f"• Cantidad de estudiantes entre 1 y 6 horas: {c7} estudiantes.")

# 8. Porcentaje más de 9 horas
c8 = (redes_est[redes_est > 9].count() / total_est) * 100
print(f"• Porcentaje de estudiantes que usan más de 9 horas: {round(c8, 2)}%")
```
