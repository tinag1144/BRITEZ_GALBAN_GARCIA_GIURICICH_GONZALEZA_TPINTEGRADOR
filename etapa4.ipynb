```python
import pandas as pd
```

```python
df = pd.read_csv("datos.csv")
horas = df['Horas de  consumo de redes sociales por día']
```

```python
frecuencia_absoluta = horas.value_counts().sort_index()
frecuencia_relativa = horas.value_counts(normalize=True).sort_index()
frecuencia_abs_acum = frecuencia_absoluta.cumsum()
frecuencia_rel_acum = frecuencia_relativa.cumsum()
```

```python
tabla_frecuencias = pd.DataFrame({
    'fi': frecuencia_absoluta,
    'Fi': frecuencia_relativa,
    'fri': frecuencia_abs_acum,
    'Fri': frecuencia_rel_acum
})
```

```python
media = horas.mean()
desviacion_est = horas.std()
```

```python
tabla_frecuencias
```

```python
print(f'Media de horas: {media: .2f}')
```

```python
print(f"Desviación estándar: {desviacion_est:.2f}")
```
